Question 10:Testing for Differences in ALT Levels Across Hepatitis Stages Define a null and alternative hypothesis to evaluate if there’s a statistically significant difference in ALT (Alanine Aminotransferase) levels across different categories in the dataset (e.g., blood donors, hepatitis patients). Choose an appropriate statistical test to assess this relationship.

```python
# Define null and alternative hypotheses
# H0: There is no difference in ALT levels across different categories.
# H1: There is a difference in ALT levels across different categories.

# Perform ANOVA test
category_column = 'Category'  # Replace with the actual column name for categories
alt_by_category = [df['ALT'][df[category_column] == cat] for cat in df[category_column].unique()]
anova_result = stats.f_oneway(*alt_by_category)
print("\nANOVA Test for ALT levels across categories:")
print(f"F-statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")

if anova_result.pvalue < 0.05:
    print("Reject the null hypothesis: Significant differences exist in ALT levels across categories.")
else:
    print("Fail to reject the null hypothesis: No significant differences in ALT levels.")
```

Question 11: Perform a Chi-Square test to assess the relationship between Category (e.g., blood donor, hepatitis status) and Sex. Interpret the test results and discuss whether there is a statistically significant association between these variables.

```python
# Contingency table
contingency_table = pd.crosstab(df['Category'], df['Sex'])

# Perform Chi-Square test
chi2_res = stats.chi2_contingency(contingency_table)
print("\nChi-Square Test for Category and Sex:")
print(f"Chi-Square Statistic: {chi2_res[0]}, p-value: {chi2_res[1]}")

if chi2_res[1] < 0.05:
    print("Reject the null hypothesis: Significant association exists between Category and Sex.")
else:
    print("Fail to reject the null hypothesis: No significant association between Category and Sex.")
```
